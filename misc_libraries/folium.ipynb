## Create Maps with Folium Library

```python
import folium

# Create a map centered at the given latitude and longitude for united states
usa_map = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

# Add a marker for Richmond, VA
folium.Marker(location=[37.5407, -77.4360], popup='Richmond, VA').add_to(usa_map)

# Add a marker for California
folium.Marker(location=[36.7783, -119.4179], popup='California').add_to(usa_map)

usa_map
```
