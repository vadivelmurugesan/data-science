## Introduction to Matplotlib
Matplotlib is an extensive library for creating static, animated, and interactive visualizations in Python. To make it versatile across multiple platforms, it offers a MATLAB-like interface.

```python
# import pyplot module
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Visualize the sex column of the Titanic Data Set

titanic_dataset = sns.load_dataset('titanic')

# Get counts of Male and Female Passengers
gender_data = titanic_dataset['sex'].value_counts()

print(f'Gender Data : {gender_data}')
```

```python
# Print type of Gender Data
print(f'\nType of Gender Data : {type(gender_data)}')

print("Below is the bar generated by the plot function.")

# Create a bar chart
gender_data.plot(kind= 'bar', title = 'Gender Distribution')
```

```python
# Labels and Titles
gender_data.plot(kind= 'bar', title = 'Gender Distribution')

# now use plt (which is an alisas name for pyplot which we have defined above while importing the module )
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')
plt.show()
```

### Customize the plot

* **color**: Sets the color of the plot.
* **alpha**: Sets the transparency level
* **grid**: Whether or not to display grid lines

```python
## Customize the plot with additional parameters
gender_data.plot(kind = 'bar', title = "Gender Distribution", color = "green", alpha = 0.9, grid= True )
```
