## HeatMaps and Correlation Matrix

### Correlation Matrix

A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation between two variables. The value is in the range of -1 to 1. If two variables have a high correlation, it means that when one variable changes, the other variable tends to change in a specific direction.

- **1** indicates a perfect positive correlation: as one variable increases, the other variable increases.
- **-1** indicates a perfect negative correlation: as one variable increases, the other variable decreases.
- **0** indicates no correlation: the variables do not have any linear relationship.

In the context of the Titanic dataset, the correlation matrix helps us understand the relationships between different numerical features such as age, fare, number of siblings/spouses aboard, etc. For example, we can see how the fare correlates with the class of the ticket or how the age correlates with the survival rate.

The correlation matrix for the Titanic dataset is calculated using the `corr()` method from pandas, which computes the pairwise correlation of columns, excluding NA/null values.

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Load Titanic dataset
titanic_df = sns.load_dataset('titanic')

corr = titanic_df.corr(numeric_only=True)
corr
```

### Heatmaps

A heatmap is a data visualization technique that shows the magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving visual cues about how the phenomenon is clustered or varies over space. Heatmaps are particularly useful for visualizing the correlation matrix of a dataset.

In the context of the Titanic dataset, a heatmap can help us visualize the correlation matrix, making it easier to see which variables are positively or negatively correlated with each other. The heatmap uses color gradients to represent the strength of the correlations, with different colors indicating different levels of correlation.

For example:
- Darker colors might represent stronger correlations (either positive or negative).
- Lighter colors might represent weaker correlations.

By visualizing the correlation matrix as a heatmap, we can quickly identify patterns and relationships between variables that might not be immediately obvious from the raw correlation matrix.

```python
# Create a heatmap
sns.heatmap(corr, annot=True)

# Show plot
plt.show()
```

```python
# Create a heatmap
sns.heatmap(corr, annot=True, cbar=True, vmin=-1, vmax=1, cmap='coolwarm')

# Show plot
plt.show()
```

```python
# Create a custom diverging color map
cmap = sns.diverging_palette(220, 20, as_cmap=True)

# Create a heatmap with the custom color map
sns.heatmap(corr, annot=True, cbar=True, vmin=-1, vmax=1, cmap=cmap)

# Show plot
plt.show()
```
