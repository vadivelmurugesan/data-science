## Seaborn

Helps to get sample datasets and build visually appealing plots effortlessly.

```python
# import library

import seaborn as sns
import matplotlib.pyplot as plt
```

### Sample DataSets

```python
# Titanic DataSets
titanic_dataset = sns.load_dataset('titanic')
print(titanic_dataset.head())
```

```python
# Newyork Taxi DataSets
taxi_dataset = sns.load_dataset('taxis')
print(taxi_dataset.head())
```

### Data Visulatization with Seaborn

Seaborn is a data visualization library built on top of matplotlib library in python.

```python
# Set aesthetics for all the plots to whitegrid.
sns.set_theme(style='whitegrid')
```

The style='whitegrid' parameter in the sns.set_theme() creates a white background with gridlines. Five other preset seaborn themes are available: darkgrid, whitegrid, dark, white, and ticks.

The other options are

* palette: Color palettes
* font: The font for all text in the plot.
* font_scale: Scale the size of the font elements.
* color_codes: If set to True, shorthand notation can be used for colors in the palette (like 'b' for blue).

```python
sns.set_theme(style='whitegrid', palette='Reds', font='Serif', font_scale=1.2)
sns.countplot(x = 'pclass', data=titanic_dataset, palette='coolwarm', hue='pclass')
plt.figure(figsize=(20, 6))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='sex', data=titanic_dataset, palette='coolwarm', hue='sex')
plt.title('Gender Count')
plt.xlabel('Gender Class')
plt.ylabel('Count')
plt.legend(title='Gender Class')
plt.xticks(rotation=90)
plt.show()
```

### Histogram

* Displays frequency distribution (shape) of a continuous or discrete data set
* Helps to visualize the distribution of a variable over a range of values.

```python
# Seaborn provides a function called histplot for creating histograms. Here's the basic syntax:

sns.histplot(data=titanic_dataset, x='age', kde=True, bins=30, color='blue')
plt.title('Age Distribution Among Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
# kde=True part is there also to draw a curve of Kernel Density Estimation (KDE) that provides a smooth estimate of our distribution.
```

### Customizing Historgram Diagram using Seaborn

* hue: To color the histogram bars for the age distribution differently depending on the passenger's gender values.

* multiple: It is used with hue to change how the different categories are displayed on the histogram. The default is "layer", but you could set it to "stack", "fill" or "dodge".

* layer: Draw one histogram per variable. Each histogram will represent a separate layer; layers will be superimposed on each other.

* stack: Draw one histogram, stacking the values of each variable on top of the other.

* fill: Draw one histogram, with the area of each filled up to the total height, the cumulating contribution of each variable (like a percentage plot, where the whole plot is 100%).

* dodge: Draw one histogram, but "dodge" them, i.e., move them slightly to the side so each contributes to the overall figure separately and all can be seen.

* palette: To change the colors used for the different categories.

* binwidth: To set the width of the bins rather than the number of bins. This can be useful for a more direct control of the granularity of the histogram.

* element: By default, the histogram is made of bars, but you could set this parameter to "step" or "poly" to change the appearance of the histogram.

```python
sns.histplot(data=titanic_dataset, x='age', hue='sex', multiple='stack', palette='pastel')
```

```python
sns.histplot(data=titanic_dataset, x='age', binwidth=1, color='purple', element='step')
```

## Bar Plots

**Helps to display and compare the number, frequency, or other measures (e.g., mean) for different categories or groups**

* hue - allows to represent an additional categorical variable by colors.
* color - allows to set a specific color for all the plot bars.
* order and hue_order - to arrange the bars in a specific order. 
* orient - to change the plot's orientation. By default, it's set to 'v' for vertical plots. You can change it to 'h' for horizontal plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# load 
titanic_dataset = sns.load_dataset('titanic')

sns.countplot(x='sex', data=titanic_dataset, hue = 'survived', palette='light:cyan', order = ['female', 'male'], orient='v').set_title('Sex and Survival Rates')

plt.show()
```

```python
# Passenger Class and Survival Rates

sns.countplot(x='pclass', data=titanic_dataset, hue = 'survived').set_title('Passenger Class and Survival Rates')

plt.show()
```

```python
# Embarked and Survival Rates

sns.countplot(x='embarked', data=titanic_dataset, hue = 'survived').set_title('Embarked and Survival Rates')

plt.show()
```
