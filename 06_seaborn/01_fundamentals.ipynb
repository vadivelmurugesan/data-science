## Seaborn

Helps to get sample datasets and build visually appealing plots effortlessly.

```python
# import library

import seaborn as sns
import matplotlib.pyplot as plt
```

### Sample DataSets

```python
# Titanic DataSets
titanic_dataset = sns.load_dataset('titanic')
print(titanic_dataset.head())
```

```python
# Newyork Taxi DataSets
taxi_dataset = sns.load_dataset('taxis')
print(taxi_dataset.head())
```

### Data Visulatization with Seaborn

Seaborn is a data visualization library built on top of matplotlib library in python.

```python
# Set aesthetics for all the plots to whitegrid.
sns.set_theme(style='whitegrid')
```

The style='whitegrid' parameter in the sns.set_theme() creates a white background with gridlines. Five other preset seaborn themes are available: darkgrid, whitegrid, dark, white, and ticks.

The other options are

* palette: Color palettes
* font: The font for all text in the plot.
* font_scale: Scale the size of the font elements.
* color_codes: If set to True, shorthand notation can be used for colors in the palette (like 'b' for blue).

```python
sns.set_theme(style='whitegrid', palette='Reds', font='Serif', font_scale=1.2)
sns.countplot(x = 'pclass', data=titanic_dataset, palette='coolwarm', hue='pclass')
plt.figure(figsize=(20, 6))
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='sex', data=titanic_dataset, palette='coolwarm', hue='sex')
plt.title('Gender Count')
plt.xlabel('Gender Class')
plt.ylabel('Count')
plt.legend(title='Gender Class')
plt.xticks(rotation=90)
plt.show()
```
