## Box Plot or Whisker Plot

Helps to display the data distribution based on number summary: the minimum, the maximum, the sample median, and the first and third quartiles. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the interquartile range.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
sns.boxplot(x='class', y='age', data=titanic)
plt.title('Boxplot of Age by Class')
plt.show()
```

* The box represents the interquartile range (i.e., 25th to 75th percentile) of the ages in each class.
* The line in the middle of the box is the median age in that class.
* The whiskers (lines extending from the box) represent the age range within 1.5 times the interquartile range above the upper and lower quartile.
* Any points beyond the whiskers can be considered outliers in the fare distribution within each class.


### Add third dimension of categorical using Hue parameter

```python
sns.boxplot(x='class', y='age', hue='survived', data=titanic)
plt.title('Age vs passenger class by survival')
plt.show()
```

### Adding Additional Parameters

```python
sns.boxplot(x='pclass', y='fare', hue='survived', data=titanic, orient='x', palette='rainbow', linewidth=0.75, linecolor='red', order=[3, 1, 2], hue_order=[1, 0], saturation=0.75, dodge=True, fliersize=5, whis=1.5)
plt.title('Fare vs Passenger Classes differentiated by Survival')
plt.show()
```
