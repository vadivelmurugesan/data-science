### Scatter Plots

A scatter plot helps to understand correlation between the variables. Each point on the plot represents an obeservation in the dataset, with its position along X and Y axes representing the value of two variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Display Scatter Plot of Age vs Fare
sns.scatterplot(x='age', y='fare', data=titanic)
plt.title("Age vs Fare")
plt.show()
```

### Compare Age vs Fare with respect to Passenger Class

Hue helps to see the data from fourth dimension. It can determine the color of data points using an additional variable.

```python
sns.scatterplot(x='age', y='fare', hue='pclass', data=titanic)
plt.title("Age vs Fare (with Passenger Class)")
plt.show()
```

### View the data in Multi Dimensional

```python
### Add further dimensions with styles based on sex and size based on fare.

sns.scatterplot(x='age', y='fare', hue='pclass', style='sex', size='fare', data=titanic, sizes=(10, 200))
plt.title("Age vs Fare (with Passenger class, Size and Fare)")
plt.show()
```

### Correlation Coefficient 

A correlation coefficient is a numerical measure of some type of correlation, meaning a statistical relationship between two variables. The variables may be two columns of a given data set of observations, often called a sample, or two components of a multivariate random variable with a known distribution. In the case of the Titanic dataset, we can calculate the correlation coefficient between the age and fare columns to determine the strength and direction of the relationship between these two variables.

* the value +1 represents an exact positive linear relationship between the two variables
* the value -1 represents an exact negative linear relationship between the two variables
* the value 0 represents no linear relationship between the two variables

```python
# Let's find correlation between all numeric variables in Titanic dataset
correlation = titanic.corr(numeric_only=True)
print(correlation)
```
